```python
from pydantic import BaseModel
```

```python
class employee_schema(BaseModel):
    name: str
    id: int
    department: str
```

```python
obj = employee_schema(name="John Doe", id=123, department="HR")
obj
```

## **Coercion VS Strict Types**

### COERCION

```python
obj3 = employee_schema(**{"name": "John Doe", "id": "101", "department": "HR"})
obj3
```

### StrictTypes

```python
from pydantic import StrictInt

class employee_schema(BaseModel):
    name: str
    id: StrictInt
    department: str
#error "id": "101"
# obj4 = employee_schema(**{"name": "John Doe", "id": "101", "department": "HR"}) 
obj4 = employee_schema(**{"name": "John Doe", "id": 101, "department": "HR"})
obj4
```
